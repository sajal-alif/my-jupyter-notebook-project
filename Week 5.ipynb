```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("gapminder.csv")
```

```python
df.dropna(inplace=True)
```

```python
df['year'] = df['year'].astype(int)
```

```python
# Group by year and calculate the mean life expectancy
life_exp_over_time = df.groupby('year')['lifeExp'].mean()

# Plot the data
plt.figure(figsize=(10, 6))
life_exp_over_time.plot(kind='line')
plt.title('Average Life Expectancy Over Time')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()
```

```python
# Filter data for a specific year (e.g., 2007)
df_2007 = df[df['year'] == 2007]

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='gdpPercap', y='lifeExp', data=df_2007, hue='continent')
plt.title('GDP per Capita vs Life Expectancy (2007)')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()
```

```python
# Group by continent and year, then sum the population
pop_growth = df.groupby(['continent', 'year'])['pop'].sum().unstack()

# Plot the data
plt.figure(figsize=(12, 8))
pop_growth.T.plot(kind='line')
plt.title('Population Growth by Continent Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.legend(title='Continent')
plt.show()
```

```python
# Filter data for 2007 and sort by GDP per capita
top_10_gdp_2007 = df_2007.sort_values(by='gdpPercap', ascending=False).head(10)

# Bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='gdpPercap', y='country', data=top_10_gdp_2007, palette='viridis')
plt.title('Top 10 Countries by GDP per Capita (2007)')
plt.xlabel('GDP per Capita')
plt.ylabel('Country')
plt.show()
```
